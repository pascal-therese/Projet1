```python
def test_mots(liste_mots,mot):
    compteur=0
    for mots in liste_mots:
        if mots==mot : compteur=compteur+1
    return compteur
             
```

```python

def dico_mots(liste_mots):
    dictionnaire={}
    for mots in liste_mots :
        if mots in dictionnaire.keys():
            dictionnaire[mots]=dictionnaire[mots]+1
        else : 
            dictionnaire[mots]=1
    return dictionnaire
```

```python
dico_mots(liste_mots)
```

```python
def dico_lettres(liste_mots):
    dico_lettres={}
    for mot in liste_mots:
        for car in mot:
            if car in dico_lettres.keys():
                dico_lettres[car]=dico_lettres[car]+1
            else : dico_lettres[car]=1
    return dico_lettres

               
```

```python
dico_lettres(liste_mots)
```
